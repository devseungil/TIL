# 한글폰트, 마이너스깨짐현상방지

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #한글폰트
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 깨지는현상해결
```

------------------------------------------------------------------

```python
x = [1,2,3]
y = [2,4,8]
plt.figure(figsize=(6,3))
plt.plot(x,y,color='r', label='무슨데이터',lw=5,marker='o',linestyle=':',markersize=10);
#lw = linewidth ls= linestyle 등 축약어 ms=markersize
plt.title('asd');
plt.xlabel('엑스라벨' ,color='r',loc='right') # x는 loc = left,right,center
plt.ylabel('와이라벨', color='b', loc='top') # y는 loc = top,bottom,center
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.legend()
plt.show()
#plt.savefig('~~~.png', dpi=100) dpi는 해상도(그림크기)
```

```python
plt.plot([-1,0,1] , [-2,0,3])
plt.show()
```

```python
plt.plot(x,y,'g^:') # 색,마커,라인스타일
```

# 텍스트삽입

```python
plt.plot(x,y,marker='o')
for idx,txt in enumerate(y):
    plt.text(x[idx],y[idx]+0.2,txt,ha='center')
            #x좌표값,y좌표값에 텍스트를 넣을거다 ha(위치)
            #x[0,1,2] = 1,2,3
            #y[0,1,2] = 2,4,8
```

# 여러 데이터 차트에 삽입

```python
days = [1,2,3]
az = [2,4,8] # (단위 : 만명) 1일부터 3일까지 아스트라 제네카 접종인구
pfizer = [5,1,3]
moderna = [1,2,5]

plt.plot(days, az, label='a' , ls=':')
plt.plot(days, pfizer, label = 'p' , ls='--')
plt.plot(days, moderna , label = 'm')
plt.xlim(1,5)
plt.ylim(1,6)
plt.legend(ncol=3) # 3개 가로로 펼쳐서
```
